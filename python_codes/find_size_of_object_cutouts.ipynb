```python
north = WCS('running_GALFIT/og_images/goodsn_3dhst.v4.0.F160W_orig_sci.fits')
south = WCS('running_GALFIT/og_images/goodss_3dhst.v4.0.F160W_orig_sci.fits')
```

```python
from runningGALFIT import get_position
#get_position(field, gid)
#return x, y, coord
```

```python
field_names = ['GN1', 'GN2', 'GN3', 'GN4', 'GN5', 'GN7', 'ERSPRIME', 'GS1', 'GS2', 'GS3', 'GS4', 'GS5']
```

```python
for i in field_names:
    
    print('\n'+str(i))
    
    #get ids
    exp_times = Table.read('/Users/rosaliaobrien/research/website/exp_times/'+i+'_exptimes.dat', format = 'ascii').to_pandas()
    ids=exp_times['id']
    
    #define fields for get_position function
    if 'N' in i:
        field = 'N'
        
    else:
        field = 'S'
    
    for index, ii in enumerate(ids):
        
        
        #only do first couple gids because they should al lbe more or less the same
        if index < 3:
        
            #define x and y position of object (in pixel space)
            x, y, coord = get_position(field, ii)

            #assuming cutouts are 100x100 pixels
            #define boundaries of cutouts
            xmax = x+50
            xmin = x-50
            ymax = y+50
            ymin = y-50

            #convert boundaries of cutouts to RA&DEC
            
            if field == 'N':
                wxmax, wymax = north.wcs_pix2world(xmax, ymax, 1)
                wxmin, wymin = north.wcs_pix2world(xmin, ymin, 1)
                
            if field == 'S':
                wxmax, wymax = south.wcs_pix2world(xmax, ymax, 1)
                wxmin, wymin = south.wcs_pix2world(xmin, ymin, 1)                

            #get cutout dimensions (in RA&DEC)
            #note: they should all be the same within 5 decimal places
            cutoutx = wxmax-wxmin
            cutouty = wymax-wymin
            
            print(str(ii)+': '+str(cutoutx)+' x '+str(cutouty)+' (deg)')
            
        else:
            pass
        
#used online converter to convert from degrees to 
```
