```python
import pandas as pd
import numpy as np
import csv

GN2data = pd.read_csv('/Users/rosaliaobrien/research/website_data/GN2.cat', sep=' ')
print(GN2data)
```

```python
def circle_pos(x_pos ,y_pos ,gidn):
    text='var gid{0} = xy({1}/sc, {2}/sc)'.format(gidn, x_pos, y_pos)
    return text
```

```python
gid = GN2data['ids']
xpos = GN2data['x']
ypos = GN2data['y']

for i, x, y in zip(gid, xpos, ypos):
        positions = circle_pos(x,np.abs(y - 6025),i)
        print(positions)
```
