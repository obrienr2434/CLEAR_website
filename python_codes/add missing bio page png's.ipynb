```python
#determine what png's we need

subfields = ['GN2']

#define variable to be png's that are needed
pngneed = []

#for each subfield
for j in subfields:
    
    #determine gid's per subfield
    df=pd.read_csv('/Users/rosaliaobrien/research/website_data/'+j+'.cat', sep=' ')
    gid=df['ids']
    
    filetypes = ['full', 'line', 'stack']
    
    for i in gid:
        for k in filetypes:
            #added a 0 to add missing bio pages that begin w/ zero
            pngneed4 = ['/Users/rosaliaobrien/research/all_png/'+str(j)+'_0'+str(i)+'.'+str(k)+'.png']
            
            #make array of all png's need from for loop (list of pngneed4's that are created in loop)
            for m in pngneed4:
                pngneed.append(m)
```

```python
pngmissing1 = []
pngmissing2 = []
pngmissing3 = []

for n in pngneed:
    if n not in allpng:
        pngmissing1.append(n)

for s in pngmissing1:
    pngmissing2 = [s.replace('/Users/rosaliaobrien/research/all_png/', '')]
    
    for q in pngmissing2:
        pngmissing3.append(q)
    
print(pngmissing3)
```

```python
#move missing files from R2_plots to all_png
for i in pngmissing3:
    for j in subfields:
        if j in i:
            try:
                shutil.move('/Users/rosaliaobrien/Desktop/R2_plots/'+j+'/'+i, "/Users/rosaliaobrien/research/all_png/missing_pngs/missing_pngs2")
            except:
                pass
```
