```python
import csv

def code(xpos, ypos, gidn, radn, rradn):
    
    gid = str(gidn)
    rrad = str(rradn)
   
    return "var circle = L.circle(gid" + gid + ", {" + "\n" + "color: 'red'," + "\n" + "fillOpacity: 0," + "\n" + "opacity: 10," + "\n" + "radius:" + rrad + "7" + "\n" + "}).addTo(map);" + "\n" + "circle.bindPopup('gid#" + gid + ": ' + '<a href=\"bio_pages/bio_page_" + gid + '.html\" target="_blank">' + " BIO PAGE</a>')" + "\n" + "circle.on('mouseover', function (e) {" + "\n" + "circlegid='" + gid + "';" + "\n" + "document.getElementById('circle_tracker').innerHTML=circlegid" + "\n" + "});" + "\n" + "circle.on('mouseout', function (e) {" + "\n" + "circlegid='Move cursor over object';" + "\n" + "document.getElementById('circle_tracker').innerHTML=circlegid" + "\n" + "});"


x = []
y = []
z = []
data = []
m = []
n = []
            
with open ('/Users/rosaliaobrien/Desktop/fields/tables/goods_south_OG/ERSPRIME_table.csv') as Data:
    #Create a way to get csv module to read column
    #Columns in ERSPRIME_table file is as follows:
    #0-gid#, 1-RA, 2-DEC, 3-x pos, 4-y pos, 5-circularized effective radius in arc seconds (uneffective)
    #6- rad given by vince, 7- log(rad/4.8)*6, 8- rad/5
    
    plots = csv.reader(Data, delimiter=',')
    for column in plots:
        x.append(float(column[3]))
        y.append(float(column[4]))
        z.append(int(column[0]))
        m.append(float(column[5])*20)
        n.append(float(column[7]))

    for i, j, k, l, p in zip(x, y, z, m, n):
        data.append(code(i, j, k, l, p))
            
print(' '.join(map(str,data))) 
```
