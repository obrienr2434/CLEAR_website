```python
def circle_pos(x_pos ,y_pos ,gidn):
    text='var gid{0} = xy({1}/3.6, {2}/3.6)'.format(gidn, x_pos, y_pos)
    return text
```

```python
gid = ERSdata['gid']
xpos = ERSdata['x_pos']
ypos = ERSdata['y_pos']

for i, x, y in zip(gid, xpos, ypos):
        positions=circle_pos(x,y,i)
        print(positions)
```
